```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 100, 101)
m0 = 1
m1 = 1.2
s0 = 50
s1 = 40
k0 = 10
k1 = 5

y0 = m0 / (1 + np.exp((s0-x)/k0))
y_s = m0 / (1 + np.exp((s1-x)/k0))
y_k = m0 / (1 + np.exp((s0-x)/k1))
y_m = m1 / (1 + np.exp((s0-x)/k0))

fig, ax = plt.subplots(figsize=(3, 3))

ax.plot(x, y0, color='k', ls='--')
ax.plot(x, y_s, color='r')
ax.plot(x, y_k, color='b')
ax.plot(x, y_m, color='g')

ax.set_ylabel('MEP amplitude')
ax.set_xlabel('stimulation intensity')

fig.savefig('/Users/mnlmrc/Library/CloudStorage/OneDrive-TheUniversityofWesternOntario/Fellowship2026/MSCA/io_curve.pdf')

plt.show()
```
