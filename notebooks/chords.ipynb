```python
import numpy as np 
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import statsmodels.formula.api as smf
from scipy.optimize import nnls
from scipy.linalg import null_space
```

# Load and prepare data

```python
path = '/cifs/diedrichsen/data/Chord_exp/EFC_all242_chords/analysis/efc1_chord.tsv'
chords = pd.read_csv(os.path.join(path), sep='\t',)
chords = chords[chords.accuracy>0].reset_index()
chords.MD = pd.to_numeric(chords.MD, errors='coerce')
chords.ET = pd.to_numeric(chords.ET, errors='coerce')
var_thresh = .9
Nf = 5
Nc = 10
col_f = [f'force_f{i + 1}' for i in range(Nf)]
col_e = [f'force_e{i + 1}' for i in range(Nf)]
col = col_f + col_e
single_finger = chords[chords.num_fingers==1].copy()
multi_finger = chords[chords.num_fingers>1].copy().reset_index(drop=True)
sns = chords.sn.unique()
N = len(sns)
n_perms = 100
n_targets = 100
theta_target = rng.uniform(0, 2 * np.pi, n_targets)
radius = 4
pos_target = radius * np.column_stack([np.cos(theta_target), np.sin(theta_target)])
```

# Helper functions

```python
rng = np.random.default_rng(0)

def dist_from_subspace(F, B):
    proj = F @ B.T @ B # component inside the subspace
    resid = F - proj # component orthogonal to it
    dist = np.linalg.norm(resid, axis=1) # norm of orthogonal component
    return dist

def angle_from_subspace(F, B):
    proj = F @ B.T @ B # component inside the subspace
    resid = F - proj # component orthogonal to it
    norm_proj = np.linalg.norm(proj, axis=1)
    norm_resid = np.linalg.norm(resid, axis=1)
    theta = np.arctan2(norm_resid, norm_proj)   # radians, in [0, π/2]
    return theta

def make_D_off(D_on, B, theta=0):
    B_perp = null_space(B).T                                  # (Nc - d, Nc)
    k = B_perp.shape[0]                                       # Nc - d
    W_perp, _ = np.linalg.qr(rng.standard_normal((k, 2)))     # (k, 2) orthonormal cols
    D_perp = W_perp.T @ B_perp                                # (2, Nc), in row(B)^perp
    D_perp *= np.linalg.norm(D_on) / np.linalg.norm(D_perp)   # match Frobenius norm
    return np.cos(theta) * D_on + np.sin(theta) * D_perp      # (2, Nc)

def make_D_on(B):
    d = B.shape[0]
    W, _ = np.linalg.qr(rng.standard_normal((d, 2)))
    D_on = W.T @ B
    return D_on

def decoder_angle(D_on, D_off):
    ax = tuple(range(D_off.ndim - D_on.ndim, D_off.ndim))   # trailing axes matching D_on
    num = np.sum(D_on * D_off, axis=ax)
    den = np.linalg.norm(D_on) * np.linalg.norm(D_off, axis=ax)
    cosang = np.clip(num / den, -1.0, 1.0)
    return np.arccos(cosang)  # radians
```

# Intrinsic force manifold and decoders definition
For each participant, I first determined the intrinsic force manifold $\mathbf{B}\in\mathbb{R}^{k\times 10}$ as the subspace spanned by the $k$ orthogonal components that explained >90% of the variance in flexion-extension force patterns during single finger movements. For each multi-finger force pattern (chord) $\mathbf{f}\in\mathbb{R}^{10}$, I determined the component lying *within* the manifold and a component *orthogonal* to it:

$$
\mathbf{f}_{\parallel} = \mathbf{f}\,\mathbf{B}^\top\mathbf{B}, \qquad
\mathbf{f}_{\perp} = \mathbf{f} - \mathbf{f}_{\parallel} = \mathbf{f}\,(\mathbf{I}-\mathbf{B}^\top\mathbf{B}).
$$

I then calculated the Euclidean distance $d$ and the angle $\theta$ from the intrinsic force manifold as:

$$
d(\mathbf{f}) = \lVert \mathbf{f}_{\perp} \rVert_2 = \bigl\lVert \mathbf{f} - \mathbf{f}\,\mathbf{B}^\top\mathbf{B} \bigr\rVert_2,
$$

$$
\theta(\mathbf{f}) = \operatorname{arctan2}\!\bigl(\lVert \mathbf{f}_{\perp} \rVert_2,\; \lVert \mathbf{f}_{\parallel} \rVert_2\bigr) \in [0, \tfrac{\pi}{2}].
$$

Next, I defined two linear decoders mapping 10-dimensional flexion-extension force patterns onto a 2-dimensional cursor: 

$$
\mathbf{pos} = \mathbf{D}\mathbf{f}
$$

where $\mathbf{pos}$ is the position of the cursor, $\mathbf{D}$ is the decoder and $\mathbf{D}$ is the input force pattern. The two decoders were designed based on the intrinsic force manifold. Specifically, $\mathbf{D_{\text{on}}}$ was defined as:

$$
\mathbf{D}_{\text{on}} = \mathbf{W}^\top\mathbf{B}\quad\in\mathbb{R}^{2\times 10}.
$$

where $\mathbf{W}\in\mathbb{R}^{k\times 2}$ is a random matrix with orthonormal columns that selects two orthonormal directions within the $k$-dimensional manifold, one per cursor axis. Because each row is a linear combination of basis vectors in the intrinsic force manifold, this decoder requires force patterns lying inside the natural single-finger subspace. We then defined $\mathbf{D}_{\perp}$ as:

$$
\mathbf{D}_{\perp}=\mathbf{W}_{\perp}^\top\mathbf{B}_{\perp}
$$

where $\mathbf{B}_{\perp}$ is the orthogonal complement of the intrinsic force manifold. Its rows therefore lie entirely *outside* the single-finger subspace. Using $\mathbf{D}_{\perp}$, we defined $\mathbf{D}_{\text{off}}$ as:

$$
\mathbf{D}_{\text{off}}(\theta) = \cos\theta\,\mathbf{D}_{\text{on}} + \sin\theta\,\mathbf{D}_{\perp}.
$$

For $\theta=\pi/2$, so $\mathbf{D}_{\text{off}}=\mathbf{D}_{\perp}$ is *fully* off-manifold, that is, reaching the same targets now requires force patterns orthogonal to the intrinsic force manifold. Finally, I calculated the force patterns required to reach some radially arranged targets based on the defined decoders $\mathbf{D}_{\text{on}}$ and $\mathbf{D}_{\text{off}}$.

```python
var_expl = np.zeros((N, Nc))
distance, angle, magnitude = [], [], []
sim = pd.DataFrame()
for s, sn in enumerate(sns):

    # single finger
    F_sf = single_finger[single_finger.sn == sn][col].to_numpy().astype(float)
    sf_mean = F_sf.mean(axis=0, keepdims=True)
    F_sf_c = F_sf - sf_mean # center

    # multi finger
    F_ch = multi_finger[multi_finger.sn == sn][col].to_numpy().astype(float)
    F_ch_c = F_ch - sf_mean
    mag_c = np.linalg.norm(F_ch_c, axis=1)
    magnitude.extend(mag_c)
    
    # PCA
    _, S, Vt = np.linalg.svd(F_sf_c, full_matrices=False)
    eigvals = S**2 / (F_sf.shape[0] - 1)
    var_expl[s] = np.cumsum(eigvals / eigvals.sum())
    d = np.where(var_expl[s] > var_thresh)[0][0] if var_expl[s, -1] > var_thresh else Nc
    B = Vt[:d]          

    # Orthogonal distance from subspace
    eucl = dist_from_subspace(F_ch_c, B)
    distance.extend(eucl)

    # Angle from subspace
    theta = angle_from_subspace(F_ch_c, B)
    angle.extend(theta)    

    # on-manifold decoder
    gains = 1 + rng.normal(0, 0.5, 2)   
    D_on = make_D_on(B)
    F_on = pos_target @ (gains[:, None] * D_on)
    mag_on = np.linalg.norm(F_on, axis=1)
    eucl_on = dist_from_subspace(F_on, B)
    angle_on = angle_from_subspace(F_on, B)
    sim = pd.concat([sim, pd.DataFrame({
      'sn': sn,
      'manifold': 'on',
      'magnitude': mag_on,
      'theta_pos': theta_target,
      'dist':     eucl_on,
      'theta':    angle_on,})])

    # off-manifold decoder
    for n in range(n_perms):
      # p = np.random.permutation(Nc)
      gains = 1 + rng.normal(0, 0.5, Nc) 
      # G = gains[:, None] * np.eye(Nc)[p]
      D_off = make_D_off(D_off, B, np.pi / 2)        
      F_off = pos_target @ (D_off * gains) # (n_perms, n_targets, Nc)
      eucl_off = dist_from_subspace(F_off, B)
      mag_off = np.linalg.norm(F_off, axis=1)
      angle_off = angle_from_subspace(F_off, B)
      sim = pd.concat([sim, pd.DataFrame({
        'sn': sn,
        'manifold': 'off',
        'magnitude': mag_off,
        'theta_pos': theta_target,
        'dist':     eucl_off,
        'theta':    angle_off,})])
    
multi_finger['magnitude'] = magnitude
multi_finger['distance'] = distance
multi_finger['angle'] = angle
```

# The distance from the intrinsic force manifold predicts chord difficulty
I used mean deviation as an index of chord difficulty see ([see Ghavampour et al., 2025](https://journals.physiology.org/doi/full/10.1152/jn.00088.2025)). Mean deviation measures the mean distance of the multi-finger force trajectory from the ideal straight path stretching from rest and the target force pattern. Typically, mean deviation improves as chords are practiced, and is larger the more active fingers are in a chord.

To understand whether the instrinsic force manifold constrained multi-finger force patterns, I fit a linear mixed-effects model to the mean deviation using either the euclidean distance or the angle from the intrinsic force manifold as a fixed-effect predictor, plus a random intercept per participant. Both the euclidean distance ($\beta=0.251$, $p<0.001$) and the angle ($\beta=0.427$, $p<0.001$) relative to the intrinsic force manifold predicted significantly the mean deviation. 

```python
md_align = smf.mixedlm("MD ~ distance", multi_finger, groups=multi_finger['sn']).fit()
md_align.summary()
```

```python
md_align = smf.mixedlm("MD ~ angle", multi_finger, groups=multi_finger['sn']).fit()
md_align.summary()
```

```python
fig, axs = plt.subplots(1, 2, sharey=True)

ax = axs[0]
sb.scatterplot(data=multi_finger, ax=ax, x='distance', y='MD', alpha=.1, color='grey', edgecolor='none', marker='.')
sb.regplot(data=multi_finger, ax=ax, x='distance', y='MD', scatter=False, color='black')
ax.set_title('Euclidean distance')
ax.set_xlabel('distance (N)')
ax.set_ylabel('mean deviation (N)')

ax = axs[1]
sb.scatterplot(data=multi_finger, ax=ax, x='angle', y='MD', alpha=.2, color='grey', edgecolor='none',marker='.')
sb.regplot(data=multi_finger, ax=ax, x='angle', y='MD', scatter=False, color='black', units='sn')
ax.set_title('Angle')
ax.set_xlabel('angle (rad)')

fig.suptitle('Distance from single finger subspace vs. chord difficulty', fontweight='bold')

plt.show()
```

# Radial targets
The radially arranged targets below where used to estimate the multi-finger force patterns required by the on and off manifold decoders ($
\mathbf{D}_{\text{on}}$ and $
\mathbf{D}_{\text{off}}$) to reach each of these locations. 

```python
fig, ax = plt.subplots(subplot_kw={'projection': 'polar'})

n_show = 10
idx = np.argsort(theta_target)[::n_targets // n_show][:n_show]

ax.scatter(theta_target, np.full(n_targets, radius), s=10, color='lightgrey',)
ax.scatter(theta_target[idx], np.full(n_show, radius), s=60, color='red',)
ax.set_ylim(0, radius + 1)
ax.set_rticks([])
#ax.set_xticks([])
ax.spines['polar'].set_visible(False)
ax.set_title('Target positions')
plt.show()
```
